```python
import pandas as pd

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer



aspects = ['camera', 'battery','display' ,'performance']

analyzer = SentimentIntensityAnalyzer()


def classify_polarity(sentiment_score):
    if sentiment_score > 0.05:
        return 'Positive'
    elif sentiment_score < -0.05:
        return 'Negative'
    else:
        return 'Neutral'


def analyze_aspect_sentiment(text, aspect):
    aspect_sentiments = []
    for sentence in str(text).split('.'):
        if aspect in sentence:
            sentiment_score = analyzer.polarity_scores(sentence)['compound']
            aspect_sentiments.append(sentiment_score)
    if aspect_sentiments:
        avg_score = sum(aspect_sentiments) / len(aspect_sentiments)
        return classify_polarity(avg_score)
    else:
        return None

for aspect in aspects:
    df[f'{aspect}_polarity'] = df['Review'].apply(lambda x: analyze_aspect_sentiment(x, aspect))

print(df)
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize


nltk.download('punkt')

aspects = ['camera', 'battery', 'display', 'performance']

analyzer = SentimentIntensityAnalyzer()

def calculate_overall_aspect_rating(aspect, df):
    aspect_sentiments = []
    for index, row in df.iterrows():
        aspect_score = 0
        aspect_sentences = sent_tokenize(str(row['Review']))
        for sentence in aspect_sentences:
            if aspect in word_tokenize(sentence.lower()):
                aspect_score += analyzer.polarity_scores(sentence)['compound']
        aspect_sentiments.append(aspect_score / len(aspect_sentences) if len(aspect_sentences) != 0 else 0)

    overall_rating = sum(aspect_sentiments) / len(df) if len(df) != 0 else 0
    normalized_rating = (overall_rating + 1) * 2.5
    return round(normalized_rating, 1)

aspect_ratings = {}
for aspect in aspects:
    aspect_ratings[aspect] = calculate_overall_aspect_rating(aspect, df)

for aspect, rating in aspect_ratings.items():
    print(f"Overall rating for {aspect}: {rating}")
```
